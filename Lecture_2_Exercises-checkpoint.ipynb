In this project, we maily analyse the stock return of BIDU Index from 2022-01-01 to 2024-12-31.

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

# Fetch Apple stock data
data = yf.download('BIDU', start='2022-01-01', end='2024-12-31')

# Select the 'Close' prices for AAPL
data_close = data[('Close', 'BIDU')]

# Calculate Log Returns
data_log_returns = np.log(data_close / data_close.shift(1))

# Drop NaN values
data_close = data_close.dropna()
data_log_returns = data_log_returns.dropna()
```

# Exercise 1

Plot the time series of the prices and log-returns of your stock. Comment on the appearance of the two plots.



```python
# Create a figure with two plots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plot Adjusted Close Prices
axs[0].plot(data_close.index, data_close, color='blue')
axs[0].set_title('BIDU Adjusted Close Prices (2022-2024)')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Price')
axs[0].grid()

# Plot Log Returns
axs[1].plot(data_log_returns.index, data_log_returns, color='green')
axs[1].set_title('BIDU Log Returns (2022-2024)')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Log Returns')
axs[1].grid()

# Adjust layout and show the plots
plt.tight_layout()
plt.savefig('bidu_prices_and_log_returns.png', dpi=300)
plt.show()
```

The adjusted close prices of BIDU are highly volatile from 2022 to 2024, fluctuating between 70 and 170, and exhibiting a decreasing trend.
The log-returns of BIDU are stationary over most of the time, except the day 2022-03-16.

# Exercise 4

Compute the 0.01 and 0.99 empirical quantiles of the log-returns. Compare them to those of the Normal distribution with mean and standard deviation estimated from the log-returns.


```python
import numpy as np
from scipy.stats import norm

# Empirical quantiles
q01_empirical = np.percentile(data_log_returns, 1)  # 0.01 quantile
q99_empirical = np.percentile(data_log_returns, 99)  # 0.99 quantile

print("Empirical 0.01 Quantile:", q01_empirical)
print("Empirical 0.99 Quantile:", q99_empirical)

# Estimated mean and standard deviation from log-returns
mean = np.mean(data_log_returns)
std = np.std(data_log_returns)

# Theoretical quantiles 
q01_normal = norm.ppf(0.01, loc=mean, scale=std)  # 0.01 quantile
q99_normal = norm.ppf(0.99, loc=mean, scale=std)  # 0.99 quantile

print("Normal Distribution 0.01 Quantile:", q01_normal)
print("Normal Distribution 0.99 Quantile:", q99_normal)
```

The difference in the 0.01 quantile is -0.00048. The empirical quantile is slightly smaller than the normal quantile;
The difference in the 0.99 quantile is 0.0094. The empirical quantile is slightly larger than the normal quantile.
This indicates that the empirical data exhibit heavier left and right tails than normal distribution.
