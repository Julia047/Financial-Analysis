```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
import statsmodels.api as sm
from statsmodels.nonparametric.smoothers_lowess import lowess

# Step 1: Fetch Apple stock data
data = yf.download('BIDU', start='2022-01-01', end='2024-12-31')

# Step 2: Calculate log returns
close = data['Close']
log_returns = np.log(close / close.shift(1)).dropna() * 100
```

# Exercise 1

Fit an AR + GARCH model to your data. You can choose it manually or use a model selection criterion. Print the summary and interpret it (with particular attention to the parameter estimates).


```python
import pandas as pd
import numpy as np
from arch import arch_model

# Initialize variables for tracking the best model
best_model = None
best_bic = np.inf
best_order = (0, 0, 0)  # (AR order, GARCH p, GARCH q)

# Search over combinations of AR(p), GARCH(p), and GARCH(q) parameters
for p in range(7):  # AR order
    for garch_p in range(1, 7):  # GARCH p (exclude 0 for GARCH)
        for garch_q in range(0, 7):
            try:
                # Fit AR(p) + GARCH(p, q) model
                model = arch_model(log_returns, mean='AR', lags=p, vol='Garch', p=garch_p, q=garch_q, dist = 't')
                result = model.fit(disp="off")

                # Track the best model based on BIC
                if result.bic < best_bic:
                    best_bic = result.bic
                    best_model = result
                    best_order = (p, garch_p, garch_q)
            except Exception as e:
                # Handle any errors during model fitting
                print(f"Error with AR({p}) + GARCH({garch_p}, {garch_q}): {e}")

# Display the best model summary
print(f"Best AR order: {best_order[0]}, Best GARCH orders: ({best_order[1]}, {best_order[2]})")
print(f"Best BIC: {best_bic}")
print(best_model.summary())
```

AR(0) suggests that past log returns do not have a significant predictive effect on current returns. The BIDU time series is more like a white noise process.

Constant (Const) = -0.1969, p-value = 0.041
The constant term is statistically significant at the 5% level, indicating that the expected daily return is slightly negative.

α₁ (ARCH effect) = 0.0358, p-value = 0.341
The high p-value (0.341) suggests that recent shocks do not strongly influence future volatility.

β₁ (GARCH effect) = 0.9499, p-value < 0.0001
Since β₁ is very close to 1 and highly significant, it suggests that volatility is extremely persistent.

# Exercise 3

Plot diagnostics of the residuals. The ACF plot of the standardized residuals, their squared and the qq-plot vs the assumed distribution. What do they tell you about your model?

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import statsmodels.api as sm
from scipy.stats import t

std_residuals = result.std_resid.dropna()  # 删除 NaN
std_residuals_squared = std_residuals ** 2
df = result.params['nu']

print(std_residuals.describe())

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# ACF of standardized residuals
plot_acf(std_residuals, lags=20, ax=axs[0])
axs[0].set_title("ACF of Standardized Residuals")

# ACF of squared standardized residuals
plot_acf(std_residuals_squared, lags=20, ax=axs[1])
axs[1].set_title("ACF of Squared Standardized Residuals")

# QQ plot of standardized residuals
sm.qqplot(std_residuals, dist=t, distargs=(df,), line="q", ax=axs[2])
axs[2].set_title("QQ Plot of Standardized Residuals (T-Distribution)")

plt.tight_layout()
plt.savefig('diagnostics.png', dpi=300)
plt.show()
```

ACF of Standardized Residuals:
Almost all autocorrelations are insignificant (inside the blue confidence band).
This suggests that the model successfully removed linear dependencies from the residuals.

ACF of Squared Standardized Residuals:
No significant autocorrelations in squared residuals, which means volatility clustering has been adequately captured by the GARCH(1,1) model.

The residuals largely follow T-distribution.

# Exercise 4

Fit a GJR GARCH model. Is there evidence of leverage?


```python
import datetime as dt
split_date = dt.datetime(2024, 1, 1)

# Fit an AR(1) + GARCH(1,1) model with t-distribution
model = arch_model(log_returns, mean='AR', lags=2, vol='Garch', p=1, q=1, dist='t')
result = model.fit(last_obs=split_date, disp="off")
forecasts = result.forecast(horizon=3, start=split_date, align='origin')
forecasts.variance[split_date:].plot()
forecasts.mean[split_date:].plot()
```
